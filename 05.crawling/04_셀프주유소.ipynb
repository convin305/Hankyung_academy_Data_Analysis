## 셀프 주유소는 정말 저렴할까?

오피넷 검색 using selenium


```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

서울시 구별 주유소 가격 정보 얻기

```python
driver = webdriver.Chrome('./chromedriver')
driver.get("https://www.opinet.co.kr/user/main/mainView.do")        # 일단 메인으로 이동
```

```python
# 싼주유소 찾기 메뉴를 찾고, 클릭
driver.find_element_by_css_selector('.ic_m1').click()
```

```python
# 서울 자치구 리스트 가져오기
region = driver.find_element_by_xpath('//*[@id="SIGUNGU_NM0"]')
gu_list = region.find_elements_by_css_selector('option')
gu_names = [gu.text for gu in gu_list]
del gu_names[0]
gu_names
```

```python
# 검색을 강남구(gu_name의 첫번째 값)으로 하기
gu = gu_names[0]
region = driver.find_element_by_css_selector("#SIGUNGU_NM0")
region.send_keys(gu)
```

```python
driver.find_element_by_xpath('//*[@id="glopopd_excel"]').click()
```

```python
# 구별로 돌아가면서 엑셀파일 받기 구별로 돌아가면서 엑셀파일 받기
for gu in gu_names:
    region = driver.find_element_by_css_selector("#SIGUNGU_NM0")
    region.send_keys(gu)
    time.sleep(1)

    driver.find_element_by_xpath('//*[@id="glopopd_excel"]').click()
    time.sleep(2)
```

```python
driver.close()
```

### 구별 휘발유 가격에 대한 데이터 정리

```python
from glob import glob
```

```python
glob("C:/Users/CPB06GameN/Downloads/지역*.xls")
```

```python
file= glob("C:/Users/CPB06GameN/Downloads/지역*.xls")[0]
df = pd.read_excel(file,header=2)
df.head()
```

```python
# 25개 파일을 모두 읽어서 하나의 데이터프레임으로 만듦.
raw = []
for file in glob("C:/Users/CPB06GameN/Downloads/지역*.xls"):
    df = pd.read_excel(file,header=2)
    raw.append(df)

station_raw = pd.concat(raw)            
```

```python
station_raw.info()
```

```python
stations = pd.DataFrame({
    '상호' : station_raw.상호,
    '주소' : station_raw.주소,
    '가격' : station_raw.휘발유,
    '셀프' : station_raw.셀프여부,
    '상표' : station_raw.상표
})
```

```python
stations.head()
```

```python
stations['구'] = [addr.split()[1] for addr in station_raw.주소]
stations.tail()
```

```python
stations['구'].unique()
```

```python
# 가격 정보가 없는 몇 개 삭제 
stations = stations[stations['가격'] != '-']
stations.info()     
```

```python
# 가격을 숫자로 변환
stations['가격'] = pd.to_numeric(stations.가격)
stations.info()
```

### 셀프 주유소는 정말 저렴할까?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline
```

```python
mpl.rcParams['axes.unicode_minus'] = False 
plt.rcParams["font.family"] = 'Malgun Gothic'
```

```python
# 셀프 여부에 따라 가격 표시한 박스플롯

stations.boxplot(column="가격",by='셀프',figsize=(12,8))
```

```python
# 상표별로 확인

plt.figure(figsize=(12,8))
sns.boxplot(x='상표',y='가격',hue='셀프',data=stations)
plt.show()
```

### 서울의 구별 휘발유 가격 확인 

```python
import json 
import folium
import warnings
warnings.simplefilter(action="ignore",category=FutureWarning)
```

```python
# 가격이 비싼 주유소 TOP5
stations.sort_values(by="가격",ascending=False).head()
```

```python
# 가격이 착한 주유소 Top 5
stations.sort_values(by="가격").head()
```

```python
gu_data = stations.groupby('구')[['가격']].mean()           # DF로 반환
gu_data
```

```python
geo_path = '../04.지도공공API/skorea_municipalities_geo_simple.json'

geo_data = json.load(open(geo_path,encoding='utf-8'))
```

```python
map = folium.Map(
        location=[37.5502, 126.982], 
        zoom_start=11, tiles='Stamen Toner'
)

map.choropleth(
    geo_data = geo_data,
    data = gu_data.가격,
    columns = [gu_data.index, gu_data.가격],
    fill_color = 'PuRd',
    key_on = 'feature.id'
)

map
```
