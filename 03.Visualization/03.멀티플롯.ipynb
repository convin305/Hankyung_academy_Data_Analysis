# 여러개의 그림 그리기

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd

%matplotlib inline
```

```python
# 한글 쓰기
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 10
```

* Figure 객체  



```python
np.random.seed(2021)
f1 = plt.figure(figsize=(10, 2))
plt.title("figure size : (10, 2)")
plt.plot(np.random.randn(100))
plt.show()
```

* Axes 객체와 subplot 명령  

ax.plot( ) : 한 그림에 여러 차트를 그릴 때, 하나를 잡아서 그리겠다는 의미. 리스트로도 가능하다. 


```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
#  그림그리기

ax1 = plt.subplot(2,1,1)
plt.title("2 Subplots")
ax1.plot(x1,y1,'yo-')
ax1.set_ylabel('Damped oscillation')

ax2 = plt.subplot(2,1,2)
ax2.plot(x2,y2,'r.-')
ax2.set_ylabel("Undamped")


plt.xlabel("Time")
plt.show()

```

```python
np.random.seed(2021)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

코드를 간결하게 쓰려면 2 * 2 같이 리스트 형태로 줘야 한다.  
위 아래 같은 그림을 그리는 코드

```python
fig, axes = plt.subplots(2, 2)      # 내가 그리고자하는 그래프의 row, column개수 

np.random.seed(2021)
axes[0, 0].plot(np.random.rand(5))          # numpy array
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)

np.random.seed(2021)
for i in range(4):
    ax = axes[i//2,i%2]     # [행인덱서, 열인덱스]
    ax.plot(np.random.rand(5))
    ax.set_title(f"axes {i+1}")
    ax.grid()

plt.tight_layout()          # 폴롯간의 간격을 자동으로 맞춰준다.
plt.show()
```

* 이중축  

DL에서 loss와 정확도를 한 그래프에 그리는 경우 유용하다. 

```python
fig, ax0 = plt.subplots()
ax1 = ax0.twinx()           # x축을 같은 걸 사용한다는 의미
ax0.set_title("2개의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
ax0.grid(False)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax1.grid(False)
ax0.set_xlabel("공유되는 x축")
plt.show()
```
