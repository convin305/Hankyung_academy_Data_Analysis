### 데이터 시각화

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd

%matplotlib inline
```

```python
# 한글 쓰기
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 10
```

* 라인플롯

```python
plt.title("Plot")
plt.plot([1, 4, 9, 16])         # X = [0,1,2,3] 가정
plt.show()
```

위와는 다르게 x값을 주니 1부터 시작하는 그래프를 볼 수 있음. 

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.title("y = x**2")
plt.grid()
plt.show()
```

* 스타일 지정

```python
plt.plot([1,2,3,4],[1,4,9,16],'rs--')
plt.title("rs-- 스타일")
plt.grid()
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("스타일 적용 예")
plt.show()
```

* 틱 설정

```python
X = np.linspace(-np.pi, np.pi, 256)
Y = np.cos(X)
plt.plot(X,Y)
plt.yticks([-1,-0.5,0,0.5,1],['Low','Mid-Low','Mid','Mid-High','High'])     # 라벨 설정도 가능 = np.linspace(-np.pi, np.pi, 256)
Y = np.cos(X)
plt.plot(X,Y)
plt.yticks([-1,-0.5,0,0.5,1],['Low','Mid-Low','Mid','Mid-High','High'])     # 라벨 설정도 가능
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
plt.title("x축과 y축의 tick label 설정")
plt.grid()
plt.show()
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
plt.grid()
plt.show()
```

* 여러개의 선 그리기

```python
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')   # 3개를 한 코드에 적용
plt.grid()
plt.show()
```

* 범례

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")         # plot안에 있는 라벨을 범례에 표시한다.
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.show()
```

* X축, Y축 라벨

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Cosine Plot")
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, ls="--", label="cosine")         
plt.plot(X, S, ls=":", label="sine")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("X,Y label을 표시한 플롯")
plt.legend(loc=2)
plt.show()
```
