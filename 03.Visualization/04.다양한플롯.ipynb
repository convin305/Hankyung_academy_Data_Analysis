## Matlib 다양한 그림 그리기  

### bar chart

```python
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

```python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.show()
```

```python
# 타이타닉의 선실별 승객 수
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
data = titanic[['class','pclass']].groupby('class').count()
             # 다시 해보기
data.plot()
plt.title("Num of passengers by class")
plt.show()
```

* 파이차트  

plt.axis('equal') : 원의 형태를 유지할 수 있도록 이 명령을 실행해 줘야 한다. 

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

* 스캐터 플랏

```python
np.random.seed(2021)
X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100)
plt.title("Scatter Plot")
plt.scatter(X, Y)
plt.show()
```

버블차트로도 표현이 가능하다.  
크기는 s, 색상은 c로 지정한다. 

```python
N = 30
np.random.seed(2021)
x = np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi * (15 * np.random.rand(N))**2
plt.title("Bubble Chart")
plt.scatter(x, y1, c=y2, s=y3)
plt.show()
```

Anscombe's quartet

```python
ans = sns.load_dataset("anscombe")
ans.info()
```

```python
ans.groupby('dataset').describe()
```

```python
ans[ans.dataset == 'I']
```

```python
fix, axes = plt.subplots(2,2)
for index, roman in enumerate(['I','II','III','IV']):
    ds = ans[ans.dataset == roman]
    ax = axes[index//2,index%2]
    ax.scatter(ds.x,ds.y)
plt.show()
```

```python
fix, axes = plt.subplots(2,2,figsize=(10,8),
                            sharex = True, sharey=True)
for index, roman in enumerate(['I','II','III','IV']):
    ds = ans[ans.dataset == roman]
    ax = axes[index//2,index%2]
    ax.plot(ds.x,ds.y,'o')                #ax.scatter(ds.x,ds.y)
    ax.plot([3,20],[4.5,13],color='r')
    ax.set_title(f'Dataset {roman}')
plt.show()
```
