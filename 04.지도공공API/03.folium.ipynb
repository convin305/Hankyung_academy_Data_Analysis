* marker 이용  
        - popup : 눌러야 이름이 뜸.  
        - tooltip :  마우스 오버하면 이름이 뜸.  
icon설정을 해서 아이콘과 그 색을 변경할 수 있음.

```python
map = folium.Map(location=[37.5598688323642,126.967109652664],zoom_start=16)
folium.Marker(location = [37.5598688323642,126.967109652664],popup='한국경제신문사').add_to(map)
map
```

```python
map = folium.Map(location=[37.5598688323642,126.967109652664],zoom_start=14)
folium.Marker(location = [37.5598688323642,126.967109652664],popup='한국경제신문사').add_to(map)
folium.Marker(location = [37.5663174209601,126.977829174031],tooltip='서울특별시청',icon=folium.Icon(color="red", icon="info-sign")).add_to(map)
map
```

* Circle : 원 형태로 표현

```python
map = folium.Map(location=[37.5598688323642,126.967109652664],zoom_start=14)
folium.Marker(location = [37.5598688323642,126.967109652664],popup='한국경제신문사').add_to(map)
folium.Marker(location = [37.5663174209601,126.977829174031],tooltip='서울특별시청',icon=folium.Icon(color="red", icon="info-sign")).add_to(map)
folium.Circle(location = [37.5641201543296,126.998009728978],popup="중구청",color='red').add_to(map)
map
```

* addr2.csv를 읽어서 지도 위에 표시하기

```python
df = pd.read_csv("addr2.csv")
df
```

```python
# 기준점을 다섯개의 위치의 평균으로 설정하기
map = folium.Map(location=[df['위도'].mean(),df['경도'].mean()],zoom_start=14)

for i in df.index:
    folium.Marker(location = [df['위도'][i],df['경도'][i]],popup = folium.Popup(df['주소'][i],min_width=220,max_width=220),tooltip=df['건물명'][i]).add_to(map)
```

```python
title_html = '<h3 align="center" style="font-size:20px"><b>5개 지역 표시</b></h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```
